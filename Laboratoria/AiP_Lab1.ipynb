```python
class Osoba():

  '''
  Klasa reprezentujaca osobe

  Atrybuty:

  Imie (str) : imie osoby
  Nazwisko (str) : nazwisko osoby
  Pesel (str) : pesel osoby
  Wiek (int): wiek osoby
  Plec (str): płeć osoby
  '''
  def __init__(self, imie, nazwisko, pesel, wiek, plec):
      self.imie(imie)
      self.nazwisko(nazwisko)
      self.pesel(pesel)
      self.wiek(wiek)
      self.plec(plec)

  def imie(self,imie):
      if not imie.isalpha():
        raise ValueError("Imię może zawierać tylko litery.")

  def nazwisko(self,nazwisko):
      if not nazwisko.isalpha():
        raise ValueError("Nazwisko może zawierać tylko litery.")

  def pesel(self, pesel):
      if not pesel.isnumeric():
        raise ValueError("Numer PESEL może zawierać tylko cyfry")
  def sprawdz_pesel(self,pesel):
    if len(pesel) != 11:
      raise ValueError("Numer PESEL musi mieć 11 cyfr")

    if not pesel.isnumeric():
      raise TypeError("Numer PESEL może składać się tylko z cyfr")

    l = [1, 3, 7, 9, 1, 3, 7, 9, 1, 3]
    sum = sum([int(pesel[i]) * l[i] for i in range(10)])
    cyfra = (10 - sum % 10) % 10

    return cyfra == int(pesel[10])
    if not sprawdz_pesel(pesel):
      raise ValueError("Numer PESEL jest nieprawidłowy.")

  def wiek(self, wiek):
      if not isinstance(wiek, int):
        raise ValueError("Wiek może zawierać tylko cyfry")
      if wiek <= 0:
        raise ValueError("Wiek nie może być ujemny.")

  def __str__(self):
      return f"{self.imie} {self.nazwisko} ({self.pesel})  {self.wiek} lat, płeć: {self.plec}"

  def __repr__(self):
    return f"Osoba(imie='{self.imie}', nazwisko='{self.nazwisko}', pesel='{self.pesel}', wiek={self.wiek}')"

  def plec(self, plec):
      if plec not in ("M", "K"):
        raise ValueError("Płeć może być tylko M lub K.")


  def zmien_nazwisko(self, nazwisko):
      self.nazwisko = nazwisko
Osoba1 = Osoba("Yahor", "Kalbasich", "12583643443434378241", 17, "M")
print(Osoba1.wiek)
print(Osoba1.Osoba)
```

```python
class KotAG:
  """ To klasa opisujaca koty. """

  def __init__(self, imie, rasa):
    self.rasa = rasa
    self.imie = imie

  @property
  def imie(self):
    return self.__imie

  @imie.setter
  def imie(self, value):
    if not isinstance(value, str):
      raise Exception("podaj napis")
    if not value.isalpha():
      raise Exception("Bez cyfr!")
    self.__imie = value

  @property
  def rasa(self):
    return self.__rasa

  @rasa.setter
  def rasa(self, val):
    if not isinstance(val, str):
      raise Exception("Podaj napis")
    if not val.isalpha():
      raise Exception("Bez cyfr!")
    self.__rasa = val





  def __str__(self):
    return f"Imie: {self.__imie}\nrasa: {self.__rasa}"
```

```python
kotag = KotAG("pazurek", "Syjamski")
print(kotag)
kotag.imie = ("Pawel")
kotag.rasa = ("British")
print(kotag)
```
